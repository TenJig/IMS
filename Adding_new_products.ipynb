Read Data from JSON

```python
import json
fd = open("record.json",'r')
r = fd.read()
fd.close()
record=json.loads(r)

```

```python
record
```

```python
prod_id = str(input("Enter product id:"))
name = str(input("Enter name:"))
pr = int(input("Enter product price:"))
qn = int(input("Enter product quantity:"))
exp = str(input("Enter the expiry:format(date/month/year)"))
ty = str(input("Enter the type of product:"))
if (prod_id in record.keys()):
  qn = qn + record[prod_id]['Quan']
record[prod_id] = {'Prod' : name, 'Prod Price' : pr, 'Quan' : qn, 'Type' : ty, 'Expiry Date' : exp}
```

```python
record
```

```python
js = json.dumps(record)
fd = open("record.json",'w')
fd.write(js)
fd.close()
```
